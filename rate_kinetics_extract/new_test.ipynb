```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd

# Get the current working directory
current_dir = os.getcwd()
print(current_dir)

# Get the root directory (assuming your script is two levels deep)
root_dir = os.path.abspath(os.path.join(current_dir, '..'))
print(root_dir)

# Add the root directory to the Python path
sys.path.insert(0, root_dir)  # Use insert(0) to prioritize this path

# Now try importing your module
from rate_kinetics_extract.RateKineticsExtract import KineticAnalysis # Only works if this name is different from the name of the file

# Now you can use your module
# Generate synthetic data for testing
time = np.linspace(0, 10, 100)  # Time points
y_initial = 0  # Baseline value
y_final = 10   # Steady state value
kon = 0.5      # Rate constant
response = y_final * (1 - np.exp(-kon * time)) + y_initial  # Synthetic response
response += np.random.normal(0, 0.5, size=len(time))  # Add noise

# Create an instance of KineticsAnalysis
kinetics = KineticAnalysis(time, response)

# Define the function to fit (baseline_steadystate_response)
def fit_function(t, y_initial, y_final, kon):
    return y_final * (1 - np.exp(-kon * t)) + y_initial

# Initial guess for the parameters
p0 = [0, 10, 0.5]  # [y_initial, y_final, kon]



# Perform the curve fit
optimal_params = kinetics.curve_fit(fit_function, p0)
print("Optimal parameters:", optimal_params)

# Plot the curve fit
kinetics.plot_curve_fit(fit_function, p0)





```
